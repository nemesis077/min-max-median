```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Function for Mean Filter
def mean_filter(image, kernel_size):
    return cv2.blur(image, (kernel_size, kernel_size))

# Function for Median Filter
def median_filter(image, kernel_size):
    return cv2.medianBlur(image, kernel_size)

# Function for Max Filter
def max_filter(image, kernel_size):
    return cv2.dilate(image, np.ones((kernel_size, kernel_size), np.uint8))

# Load an image (grayscale or color)
image = cv2.imread('/content/diksha.jpg')

# Apply Mean Filter
mean_filtered = mean_filter(image, 5)

# Apply Median Filter
median_filtered = median_filter(image, 5)

# Apply Max Filter
max_filtered = max_filter(image, 5)

# Display the results
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1), plt.imshow(image), plt.title('Original Image'), plt.axis('off')
plt.subplot(2, 2, 2), plt.imshow(mean_filtered), plt.title('Mean Filtered'), plt.axis('off')
plt.subplot(2, 2, 3), plt.imshow(median_filtered), plt.title('Median Filtered'), plt.axis('off')
plt.subplot(2, 2, 4), plt.imshow(max_filtered), plt.title('Max Filtered'), plt.axis('off')
plt.show()
```
